```python
import numpy as np
from oect_transfer import Transfer

# 示例数据：栅极电压 (Vg) 与漏电流 (Id)
# 加载csv文件
data = np.loadtxt('data/7_transfer.csv', delimiter=',', skiprows=1)
vg = data[:, 0]  # 假设第一列是栅极电压
id = data[:, 1]  # 假设第二列是漏电流

# 创建 Transfer 对象
transfer = Transfer(vg, id, device_type="N")

# 获取计算参数
print(f"最大跨导: {transfer.absgm_max.forward} S")
print(f"最大跨导: {transfer.absgm_max.forward} S")
print(f"最大跨导: {transfer.absgm_max.reverse} S")
print(f"阈值电压: {transfer.Von.forward} V")
print(f"阈值电压: {transfer.Von.reverse} V")

print(f"最大电流: {transfer.absI_max.raw} A")
```
